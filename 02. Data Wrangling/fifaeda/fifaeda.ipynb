# Зависит ли Стоимость игрока стоимость (Value) игрока от его "финтов" (Skill Moves)? 

```python
def value_converter(n):
    if n[-1] == 'M':
        mult = 1000000
    elif n[-1] == 'K':
        mult = 1000
    else:
        return float(n[1:])
    return float(n[1:-1]) * mult


df_skill_to_value = df[['Value', 'Skill Moves']]
df_skill_to_value['Value'] = df_skill_to_value['Value'].apply(value_converter)
df_value_to_skill = df_skill_to_value.groupby('Value', as_index=False).mean()
df_skill_to_value = df_skill_to_value.groupby('Skill Moves', as_index=False).median()

fig = px.bar(df_skill_to_value, x="Skill Moves", y='Value')
fig.update_layout(yaxis_title='Стоимость',
                  xaxis_title='Финты (N)',
                  title='Медианная стоимость игрока с N финтов')
fig1 = px.line(df_value_to_skill,
               x='Value', y='Skill Moves')
fig1.update_layout(yaxis_title='Финты',
                  xaxis_title='Стоимость',
                  title='Среднее кол-во финтов у игрока в зависимости от стоимости')
fig.show(renderer='colab')
fig1.show(renderer='colab')
```

## Количество финтов по стоимости предсказать с увеличением всё труднее, дальше 40М зависимостей не наблюдается на [2], несмотря на то, что по [1] видно, что в среднем игроки с большим кол-вом финтов стоят дороже (По этой же причине для [1] выбраны медианные значения, на средних "выбросы" заставляли игроков с одним финтом в среднем стоить больше игроков с двумя финтами). То есть если искать игрока по количеству финтов, можно ожидать увеличения стоимости, но работу принципа "дороже значит лучше" гарантировать нельзя.

# Есть ли зависимость скорости (SprintSpeed) от веса и роста? А от национальности?  

## Да, зависимости есть. Средняя скорость выше у тех, кто меньше весит и меньше растёт ([1], [2]). Также можно составить топ средней скорости среди стран ([3]).

```python
def height_converter(h):
    h = h.split("'")
    return int(h[0]) * 12 + int(h[1])

def weight_converter(w):
    return int(w[:-3])

df_converted_height = df[['SprintSpeed', 'Height']].dropna()
df_converted_height['Height'] = df_converted_height['Height'].apply(height_converter)

df_converted_weight = df[['SprintSpeed', 'Weight']].dropna()
df_converted_weight['Weight'] = df_converted_weight['Weight'].apply(weight_converter)

fig1 = px.histogram(df_converted_height.sort_values('Height'),
                    x="Height",
                    y="SprintSpeed",
                    histfunc="avg",
                    text_auto=True)
fig1.update_layout(title='Зависимость средней скорости от роста',
                   xaxis_title='Рост (в дюймах)',
                   yaxis_title='Средняя скорость')

fig2 = px.histogram(df_converted_weight.sort_values('Weight'),
                    x="Weight",
                    y="SprintSpeed",
                    histfunc="avg",
                    text_auto=True)
fig2.update_layout(title='Зависимость средней скорости от веса',
                   xaxis_title='Вес (в фунтах)',
                   yaxis_title='Средняя скорость')

df_nat_to_mean_speed = df[['Nationality', 'SprintSpeed']].groupby('Nationality', as_index=False).mean()
fig3 = px.histogram(df_nat_to_mean_speed.sort_values('SprintSpeed'),
                    x="Nationality",
                    y="SprintSpeed",
                    histfunc="avg",
                    text_auto=True)
fig3.update_layout(title='Топ средней скорости среди стран',
                   xaxis_title='Национальность',
                   yaxis_title='Средняя скорость')

fig1.show(renderer='colab')
fig2.show(renderer='colab')
fig3.show(renderer='colab')
```

# Есть ли зависимость между продолжительностью контракта игрока и его стоимость? 

### Да, зависимость есть, её можно наблюдать на столбчатой диаграмме:

```python
def join_converter(date):
    return int(date.split()[-1])


def CVU_converter(CVU_year):
    return int(CVU_year)


def value_converter(n):
    if n[-1] == 'M':
        mult = 1000000
    elif n[-1] == 'K':
        mult = 1000
    else:
        return float(n[1:])
    return float(n[1:-1]) * mult


df_con_len = df[['Joined', 'Contract Valid Until', 'Value']].copy().dropna()

df_con_len['Joined'] = df_con_len['Joined'].apply(join_converter)
df_con_len['Contract Valid Until'] = df_con_len['Contract Valid Until'].apply(CVU_converter)
df_con_len['Value'] = df_con_len['Value'].apply(value_converter)

df_con_len['CLength'] = df_con_len['Contract Valid Until'] - df_con_len['Joined']
df_con_len = df_con_len.groupby('CLength', as_index=False).median()

fig = px.bar(df_con_len, x='CLength', y='Value')
fig.update_layout(xaxis_title='Длительность контракта (в годах)',
                  yaxis_title='Медианная стоимость игрока',
                  title='Зависимость стоимости игрока от длительности контракта')
fig.show(renderer='colab')
```

```python
def wage_converter(n):
    if n[-1] == 'M':
        mult = 1000000
    elif n[-1] == 'K':
        mult = 1000
    else:
        return float(n[1:])
    return float(n[1:-1]) * mult


df_wage_oa = df[['Wage', 'Overall']].dropna()
df_wage_oa['Wage'] = df_wage_oa['Wage'].apply(wage_converter)
fig = px.histogram(df_wage_oa, x='Overall', y='Wage', histfunc='avg', nbins=5)
fig.update_layout(title='Средняя зарплата в зависимости от рейтинга',
                  xaxis_title='Рейтинг',
                  yaxis_title='Средняя зарплата')
fig2 = px.histogram(df_wage_oa, x='Overall', y='Wage', histfunc='avg')
fig2.update_layout(title='Средняя зарплата в зависимости от рейтинга, но поточнее',
                  xaxis_title='Рейтинг',
                  yaxis_title='Средняя зарплата')

fig.show(renderer='colab')
fig2.show(renderer='colab')
```
